```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load cleaned data
df = pd.read_csv('./data/listing_primary.csv')
df.head()
```

```python
bed_df = df['beds'].value_counts().rename_axis('beds').reset_index(name='count')
bed_df['beds'] = bed_df['beds'].astype(int)
bed_df
```

```python
df_rel = df[['beds', 'price']]
type(df_rel)
```

```python
area_df = df.groupby('neighbourhood', as_index=False)['price'].agg([np.mean, np.median, np.size])
area_df
```
