```python
import geopandas as gpd
import pandas as pd
```

```python
geo_df = gpd.read_file('Neighborhood_Map_Atlas_Neighborhoods.geojson')
geo_df.head()
```

```python
geo_df = geo_df[["L_HOOD", "S_HOOD", "geometry"]]
geo_df.head()
```

```python
df = pd.read_csv('Seattle Crime Data.csv')
df.head()
```

```python
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
seattleCrime_gdf = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")
seattleCrime_gdf.head()
```

```python
seattleCrime_gdf = gpd.sjoin(seattleCrime_gdf, geo_df, how="left", op='within')
seattleCrime_gdf.head()
```

```python
na_counts =seattleCrime_gdf.isna().sum()
na_counts
```

```python
seattleCrime_gdf= seattleCrime_gdf.dropna(axis='rows')
seattleCrime_gdf.to_csv('seattleCriemWithArea.csv', index=False)
```
