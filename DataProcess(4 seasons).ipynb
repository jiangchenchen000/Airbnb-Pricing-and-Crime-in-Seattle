```python
import geopandas as gpd
import pandas as pd
import re
pd.set_option('display.max_rows', None) 
pd.set_option('display.max_columns', None) 
```

```python
geo_df = gpd.read_file('Neighborhood_Map_Atlas_Neighborhoods.geojson')
geo_df.head()
```

```python
df1 = pd.read_csv('listings1.csv')
df2 = pd.read_csv('listings2.csv')
df3 = pd.read_csv('listings3.csv')
df4 = pd.read_csv('listings4.csv')
df = pd.concat([df1, df2, df3,df4], ignore_index=True)
df.head()
```

```python
selected_columns = ['id', 'name', 'neighbourhood_cleansed','latitude','longitude','property_type','room_type','accommodates',
                   'bathrooms_text','bedrooms','beds','price','availability_90','number_of_reviews','review_scores_rating',
                    'review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication',
                   'review_scores_location','review_scores_value'] 
df = df[selected_columns]
df.head()
```

```python
df_availability = (df.groupby('id')['availability_90'].sum().reset_index(name='availability in one year'))
unique_listing_info = df.drop_duplicates('id').drop('availability_90', axis=1)
df = pd.merge(df_availability, unique_listing_info, on='id')
df.head()
```

```python
bedrooms_pattern = r'(\d+)\s+bedroom'
bed_pattern = r'(\d+)\s+bed'

for index, row in df.iterrows():
    if pd.isna(row['bedrooms']):
        if 'studio' in row['name'].lower():
            df.loc[index, 'bedrooms'] = 0
        else:
            bedrooms_match = re.search(bedrooms_pattern, row['name'])
            if bedrooms_match:
                df.loc[index, 'bedrooms'] = int(bedrooms_match.group(1))

    if pd.isna(row['beds']):
        bed_match = re.search(bed_pattern, row['name'])
        if bed_match:
            df.loc[index, 'beds'] = int(bed_match.group(1))
```

```python
unique_bathroom_types = df['bathrooms_text'].unique()
print(unique_bathroom_types)
```

```python
import re

df['bathroom_count'] = df['bathrooms_text'].apply(lambda x: float(re.search(r'(\d+\.?\d*)', x).group(1)) if isinstance(x, str) and re.search(r'(\d+\.?\d*)', x) else 0)

def extract_bathroom_type(text):
    if not isinstance(text, str):
        return 'unknown' 
    if 'shared' in text:
        return 'shared'
    elif 'private' in text:
        return 'private'
    elif 'half-bath' in text:
        return 'half-bath'
    else:
        return 'standard'

df['bathroom_type'] = df['bathrooms_text'].apply(extract_bathroom_type)
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
missing_percentage = df.isna().sum() / len(df) * 100
print(missing_percentage)
```

```python
df = df.dropna(subset=['price','bedrooms','beds'])
```

```python
missing_percentage = df.isna().sum() / len(df) * 100
print(missing_percentage)
```
